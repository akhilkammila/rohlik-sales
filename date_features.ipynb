Clever features
- days until next shop closing (maybe people will buy now, if it is closed tmrw)
- day after closing, day before closing (day after/before holiday as well)

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
DIRECTORY = './data/'
calendar = pd.read_csv(DIRECTORY + 'calendar.csv', parse_dates=['date'])
```

```python
calendar['date'].nunique() * calendar['warehouse'].nunique() - len(calendar)
```

```python
frankfurt = calendar[calendar['warehouse'] == 'Frankfurt_1'].sort_values(by='date')
frankfurt['next_holiday_date'] = frankfurt.loc[frankfurt['holiday'] == 1, 'date'].shift(-1)
frankfurt['next_holiday_date'] = frankfurt['next_holiday_date'].bfill()
frankfurt.iloc[70:100]
```
