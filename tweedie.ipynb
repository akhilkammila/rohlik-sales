```python
import pandas as pd
import numpy as np

# Sample series with a NaN value
data = [1, np.nan, 3, 4, 5, 6, 7, np.nan, 8, 8, 19, 101, 10]
s = pd.Series(data)

# Rolling mean with a window of 3 (default min_periods=3)
# rolling_mean_default = s.rolling(window=3).mean()
# print("Default (min_periods=3):")
# print(rolling_mean_default)

# Rolling mean with a custom min_periods=1 (computes even with fewer valid entries)
rolling_mean_min1 = s.rolling(window=3, min_periods=1).mean()
print("\nCustom (min_periods=1):")
print(rolling_mean_min1)
```
